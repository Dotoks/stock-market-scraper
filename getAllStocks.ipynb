```python
import requests
import json
import pandas as pd
import tabulate
```

```python
url = "https://twelve-data1.p.rapidapi.com/stocks"
querystring = {"exchange":"NASDAQ","format":"json"}
headers = {
	"X-RapidAPI-Key": "8382bfc35bmshb1fc660c03ed003p13adadjsna9bdc9247da1",
	"X-RapidAPI-Host": "twelve-data1.p.rapidapi.com"
}
```

```python
response = requests.get(url, headers=headers, params=querystring)
```

```python
stockMarketData = response.json()
parsedStockMarketData = json.dumps(stockMarketData, indent=4)
parsed = pd.read_json(parsedStockMarketData)
s = pd.Series(parsed.data)
```

```python
stocks = pd.DataFrame()
for stock in s:
   tempStock = pd.DataFrame([stock.values()], columns=stock.keys())
   stocks = pd.concat([stocks, tempStock], ignore_index=True)
stocksTable = tabulate.tabulate(stocks, headers="keys", tablefmt="psql")
print(stocksTable)
```
